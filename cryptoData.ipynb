# Importing the required libraries 

```python

import pandas as pd # for data analysis 
import mplfinance as mpl # for plotting  

# importing the coingeek api
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

# Defining fucntion to get the OHLC data of any Coin in the respective Currency and Days

```python
# here it takes input in c, incurr & d.
# c = coin data required
# incurr = with respect to which currency
# d = of how many days 

def getOHLC(c,incurr,d):
    # fetching dtaa using api and storing to a dataframe
    s = cg.get_coin_ohlc_by_id(c,vs_currency=incurr,days=d) 
    df = pd.DataFrame(s,columns =['Date','Open','High','Low','Close'])
    # Converting the timestamp to the datetime format 
    df['Date'] = df['Date'].values.astype(dtype='datetime64[ms]')
    # Setting the Date column to index in order to further plot ohlc chart. 
    df = df.set_index('Date')

    return df
```

# Defining the fuction to plot the Candelstick graph 

```python
def plotohlc(df):
    
    # Uing mplfinance to create graph

    mpl.plot(
        df, # input dataframe
        type="candle", 
        mav =(3,6,9),
        ylabel='Price ('+curr+')',
        ylabel_lower='',
        datetime_format='%b %d',
        title = coin +' price',  
        style='yahoo')

```

# 1 . Fetching and plotting price of Ethereum (in USD)

```python
# giving input to get the data 

coin = 'ethereum' 
curr = 'usd'
days  = '180'
```

```python
# Dataframe of etherium prices in usd 

df1 = getOHLC(coin,curr,days)
df1
```

```python
# Candelstick graph

g1 = plotohlc(df1)
g1
```
